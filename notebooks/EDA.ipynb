```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df=pd.read_csv('true_cost_fast_fashion.csv')
df
```

```python
df.info()
```

```python
pivot=df.pivot_table(index='Year',
                                columns='Brand',
                                values='Monthly_Production_Tonnes',
                                aggfunc='sum').reset_index()
pivot
```

```python
fig_prod = px.line(pivot, x="Year", y=["Forever 21", "H&M", "Zara", "Uniqlo","Shein"], 
                   title="Producción mensual por marca a lo largo del tiempo")
fig_prod
```

```python
selected_countries = ['USA', 'UK', 'Germany']
selected_brands = ['H&M', 'Zara', 'Uniqlo']
selected_years = [2010, 2020]

# Filter the DataFrame based on selected countries, brands, and years

filtered_df = df[
    (df['Country'].isin(selected_countries)) & 
    (df['Brand'].isin(selected_brands)) &
    (df['Year'].between(selected_years[0], selected_years[1]))]

filtered_df
```
